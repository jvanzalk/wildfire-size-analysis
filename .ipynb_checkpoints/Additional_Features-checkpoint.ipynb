```python
import pandas as pd
import numpy as np
import geopandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import resulting dataframe from feature engineering process
df = pd.read_csv('Data/fires_cleaned.csv')
df.head()
```

```python
# Turn data into geospatial dataframe
gdf = geopandas.GeoDataFrame(
    df, geometry=geopandas.points_from_xy(df.long, df.lat))
gdf.head()
```

### Biogeoclimatic Ecosystems

```python
# Import GeoJSON of Biogeoclimatic Ecosystems (classification system based on types of vegitation, soil, and other factors)
# Source: https://catalogue.data.gov.bc.ca/dataset/bec-map#edc-pow
bec = geopandas.read_file('Data/BEC_BIOGEOCLIMATIC_POLY.geojson')
bec.head()
```

```python
#Check for missing data
sns.heatmap(bec.isnull(), cbar=False)
```

```python
# Remove unused columns
bec = bec[['OBJECTID','ZONE','ZONE_NAME','geometry']]
bec.head()
```

```python
# Determine coordinate reference systems
print('gdf: ', gdf.crs)
print('bec: ', bec.crs)
```

```python
# Set wildfires crs to match Biogeoclimatic Ecosystems
gdf.crs = {'init': 'epsg:4326'}
```

```python
# Perform a spatial join to get determine what zone each forest fire was in
# join options: https://shapely.readthedocs.io/en/latest/manual.html#binary-predicates
gdf = geopandas.sjoin(gdf, bec, how="left", op='intersects')
gdf.head()
```

```python
# Check percent of fires that have a zone now
len(gdf[gdf['ZONE'].notnull()])/len(gdf)*100
```

```python
# Remove fires without a zone
gdf = gdf[gdf['ZONE'].notnull()]
```

### Major Roads

```python
# Import GeoJSON of major roads
# Source: https://catalogue.data.gov.bc.ca/dataset/7-5m-major-roads-the-atlas-of-canada-base-maps-for-bc#edc-pow
roads = geopandas.read_file('Data/DBM_BC_7H_MIL_ROADS_LINE.geojson')
roads.head()
```

```python
# Remove columns
roads = roads[['OBJECTID','geometry']]
roads.head()
```

```python
# Calculate the minimum distance from fires to roads
def min_distance(point, lines):
    return lines.distance(point).min()

gdf['min_dist_to_roads'] = gdf.geometry.apply(min_distance, args=(roads['geometry'],))
gdf.head()
```
